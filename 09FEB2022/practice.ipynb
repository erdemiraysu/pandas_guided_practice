### Simple Plotting

```python
# Survivors.
df['survived'].value_counts(normalize=True).plot(kind='bar')
```

```python
# Filtering
male_ages = df[df['sex'] == 'male']['age']
female_ages = df[df['sex'] == 'female']['age']
```

```python
# Histogram showing male / female ages.
## LONGHAND: matplotlib and pandas filtering.
fig, ax = plt.subplots()
ax.hist(female_ages, alpha=0.5)
ax.hist(male_ages, alpha=0.5);
```

```python
## Shorthand: seaborn.
sns.histplot(data=df, x='age', hue='sex');
```

```python
# Most popular `embark_town`.
df['embark_town'].value_counts().plot(kind='bar')
```
